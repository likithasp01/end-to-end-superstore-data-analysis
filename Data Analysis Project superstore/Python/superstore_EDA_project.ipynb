```python
sns.histplot(df['Sales'])
plt.title('Sales Distribution')
plt.show()
```

```python
sns.countplot(data=df, x='Category')
plt.title('Products by Category')
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='barh')
plt.title('Profit by Region')
plt.show()
```
